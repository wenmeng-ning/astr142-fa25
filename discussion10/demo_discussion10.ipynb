```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
import matplotlib.text as text
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
circle = patches.Circle((0.3, 0.6), radius=0.2, edgecolor='blue', facecolor='skyblue', lw=2)
ax.add_patch(circle)
square = patches.Rectangle((0.5, 0.2), 0.2, 0.2, edgecolor='red', facecolor='lightcoral', lw=2)
ax.add_patch(square)
txt = text.Text(0.7,0.8,'Text',font='Times',fontsize=15)
ax.add_artist(txt)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label='sin(x)')
# data coords
ax.text(3.75, 0.70, 'Data Coordinates', fontsize=12, 
        color='purple')
# axes fraction
ax.text(0.85, 0.1, 'Axes Fraction', fontsize=14, 
        color='orange', ha='center', transform=ax.transAxes)
# figure fraction
fig.text(0.2, 0.9, 'Figure Fraction', fontsize=16, color='brown', ha='center')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()
x = np.linspace(0, 10, 100)
ys = [np.sin(x), np.cos(x), np.sin(x)+np.cos(x), np.sin(x/2)+np.cos(2*x)]
labels = ['sin(x)','cos(x)','sin(x)+cos(x)','sin(x/2+cos(2x))']
for i,ax in enumerate(axes):
    ax.plot(x,ys[i])
    ax.set_title(labels[i])
    ax.text(0.02,0.02,'Fig {}: {}'.format(i,labels[i]),transform=ax.transAxes)
plt.show()
```

```python
data = np.random.rand(10, 10) * 100
fig, ax1 = plt.subplots(figsize=(8, 6))
extent = [-0.5, 10.5, -0.5, 10.5]
cax = ax1.imshow(data, cmap='plasma', extent=extent, vmax=90)
cbar = fig.colorbar(cax, ax=ax1)
cbar.set_ticks([10, 50, 90])
cbar.set_label('Intensity')
plt.show()
```
